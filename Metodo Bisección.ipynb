Metodo de la Bisecci√≥n

$$f(x)=-2 + 7x - 5x^{2} + 6x^{3}$$

1: La funci√≥n $f(x)= -2 + 7x - 5x^{2} + 6x^{3}$ pasa por cero.

2: Tomamos un intervalo que contiene el cero, como $[‚àí2,2]$.

3: El m√©todo de bisecci√≥n encuentra la ra√≠z $ùë•=0$, con precisi√≥n dentro de la tolerancia.

4: Referencia https://lunaipi95.wordpress.com/2015/03/18/practica3-metodo-de-biseccion/

5. Resultado:https://lunaipi95.wordpress.com/wp-content/uploads/2015/03/a-7.jpg

```python
import numpy as np
```

```python
def f(x):
    return -2 + 7*x - 5*x**2 + 6*x**3

def biseccion(a, b, tol):
    if f(a) * f(b) >= 0:
        print("El intervalo no es v√°lido. La funci√≥n debe cambiar de signo en [a, b].")
        return None
    
    iteraciones = 0
    historial = []
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        historial.append((iteraciones, a, b, c, f(c)))
        
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteraciones += 1
    
    raiz = (a + b) / 2
    print(f"Ra√≠z aproximada: {raiz} despu√©s de {iteraciones} iteraciones")
    print("Iteraciones:")
    print("Iter |    a     |    b     |    c     |  f(c)")
    for it in historial:
        print(f"{it[0]:>4} | {it[1]:>8.5f} | {it[2]:>8.5f} | {it[3]:>8.5f} | {it[4]:>8.5f}")
    
    return raiz

# Definir intervalo y tolerancia
a = -2
b = 2
tolerancia = 1e-6   # Convergencia Criter√≠o 

# Llamar a la funci√≥n de bisecci√≥n
biseccion(a, b, tolerancia)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n f(x)
def f(x):
    return -2 + 7*x - 5*x**2 + 6*x**3

# Definir el intervalo de x
x = np.linspace(-2, 2, 1000)  # Genera 1000 puntos entre -2 y 2
y = f(x)  # Eval√∫a f(x) para cada valor de x

# Graficar la funci√≥n
plt.plot(x, y, label=r'$f(x) = -2 + 7x - 5x^2 + 6x^3$', color='blue')

# A√±adir l√≠neas horizontales y verticales en el eje X y Y
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Se√±alar la intersecci√≥n en x = 0
x_interseccion = 0
y_interseccion = f(x_interseccion)
plt.plot(x_interseccion, y_interseccion, 'ro')  # Marcamos el punto de intersecci√≥n con un punto rojo

# T√≠tulo y etiquetas
plt.title("Gr√°fica de la funci√≥n f(x) con intersecci√≥n en x = 0")
plt.xlabel("x")
plt.ylabel("f(x)")

# A√±adir leyenda
plt.legend()

# Mostrar la gr√°fica
plt.grid(True)
plt.show()
```
